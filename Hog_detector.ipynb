```python
from skimage.feature import hog
from cv2 import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances
```

```python
img = plt.imread("Volto1.jpg")
img2 = plt.imread("Volto2.jfif")
img = cv2.resize(img, (128,128))
img2 = cv2.resize(img2, (128,128))
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize= (10,10))
ax1.imshow(img)
ax1.axis("off")
ax2.axis("off")
ax2.imshow(img2)
print(img.shape)
print(img2.shape)
```

```python

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,10))
imgGray1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
imgGray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
ax1.imshow(imgGray1, cmap="gray")
ax2.imshow(imgGray2, cmap="gray")
```

```python

fd, hog_image = hog(imgGray1, orientations=9, pixels_per_cell=(8, 8),
                	cells_per_block=(2, 2), visualize=True, multichannel=False)
plt.axis("off")
plt.imshow(hog_image, cmap="gray")
```

```python
fd, hog_image1 = hog(imgGray2, orientations=9, pixels_per_cell=(8, 8),
                	cells_per_block=(2, 2), visualize=True, multichannel=False)
plt.axis("off")
plt.imshow(hog_image1, cmap="gray")
```

```python
cos_sim = cosine_similarity(hog_image1.reshape(1,-1), hog_image.reshape(1,-1))
print(cos_sim)
```
