```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img=mpimg.imread('burano.png')
imgplot = plt.imshow(img)
plt.show(plt.xticks([]), plt.yticks([])) 
```

#Now we have to convert the image in gray_scale to perform an edge detection

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(imgGray, cmap="gray")
plt.show(plt.xticks([]), plt.yticks([]))
```

#Now i compute the gradient magnitude in both cases

```python
grX = np.array([[-1,0,1], [-1,0,1], [-1,0,1]])
grY = np.array([[-1,-1,-1], [0,0,0], [1,1,1]])

prewittX = cv2.filter2D(imgGray, -1, grX)
prewittY = cv2.filter2D(imgGray, -1, grY)

plt.imshow(prewittX, cmap="gray")
plt.title("Prewitt X gray scale")
plt.show(plt.xticks([]), plt.yticks([]))

plt.imshow(prewittY, cmap="gray")
plt.title("Prewitt Y gray scale")
plt.show(plt.xticks([]), plt.yticks([]))
```

#Apply gradient to Prewitt filter

```python
gradX = cv2.magnitude(prewittX, prewittX)
plt.imshow(gradX, cmap="gray")
plt.title("Gradient on X whit Prewitt filter")
plt.show(plt.xticks([]), plt.yticks([]))

gradY = cv2.magnitude(prewittY, prewittY)
plt.imshow(gradY, cmap="gray")
plt.title("Gradient on Y whit Prewitt filter")
plt.show(plt.xticks([]), plt.yticks([]))
```

#Apply the Sobel filter

```python
sobelX = cv2.Sobel(imgGray, cv2.CV_64F , dx=1, dy=0, ksize=3)
sobelY = cv2.Sobel(imgGray, cv2.CV_64F , dx=0, dy=1, ksize=3)

plt.imshow(sobelX, cmap="gray")
plt.title("Sobel X filter")
plt.show(plt.xticks([]), plt.yticks([]))

plt.imshow(sobelY, cmap="gray")
plt.title("Sobel Y filter")
plt.show(plt.xticks([]), plt.yticks([]))
```

#Apply gradient to sobel filters

```python
gradSobelX=cv2.magnitude(sobelX, sobelX)
plt.imshow(gradSobelX, cmap="gray")
plt.title("Gradient on X whit Sobel filter")
plt.show(plt.xticks([]), plt.yticks([]))


gradSobelY=cv2.magnitude(sobelY, sobelY)
plt.imshow(gradSobelY, cmap="gray")
plt.title("Gradient on Y whit Sobel filter")
plt.show(plt.xticks([]), plt.yticks([]))
```

#Now i compute the gradient magnitude in both cases

```python
gradPrewitt = cv2.magnitude(prewittX, prewittY)
plt.imshow(gradPrewitt, cmap="gray")
plt.title("Gradient magnitude in Prewitt case")
plt.show(plt.xticks([]), plt.yticks([]))

gradSobel = cv2.magnitude(sobelX, sobelY)
plt.imshow(gradSobel, cmap="gray")
plt.title("Gradient magnitude in Sobel case")
plt.show(plt.xticks([]), plt.yticks([]))
```

#Now apply the thresholding features to both gradient images

```python
threShoPrewitt, resThresPrewitt = cv2.threshold(gradPrewitt,1,255, cv2.THRESH_BINARY)
threShoSobel, resThreshoSobel = cv2.threshold(gradSobel,1,255, cv2.THRESH_BINARY)

plt.imshow(resThresPrewitt, cmap="gray")
plt.title("Threshold at Prewitt image")
plt.show(plt.xticks([]),plt.yticks([]))

plt.imshow(resThreshoSobel, cmap="gray")
plt.title("Threshold at Sobel image")
plt.show(plt.xticks([]),plt.yticks([]))

```

#Compute difference between the two thresholding

```python
image = np.asarray(resThresPrewitt, np.float64)
image1 = np.asarray(resThreshoSobel, np.float64)

difference = cv2.subtract(image, image1)
difference1 = 255 - difference
plt.imshow(difference1, cmap="gray")
plt.show(plt.xticks([]), plt.yticks([]))
```

#Apply the Canny edge detector to the original image

```python
edges = cv2.Canny(img, 300, 400)

plt.imshow(edges, cmap="gray")
plt.title("Canny detector 300-400")
plt.show(plt.xticks([]), plt.yticks([]))

edges = cv2.Canny(img, 300, 400)

plt.imshow(edges, cmap="gray")
plt.title("Canny detector 30-50")
plt.show(plt.xticks([]), plt.yticks([]))
edges = cv2.Canny(img, 50, 100)

plt.imshow(edges, cmap="gray")
plt.title("Canny detector 100-250")
plt.show(plt.xticks([]), plt.yticks([]))
```
