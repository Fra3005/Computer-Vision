```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('Montagna.jpg')
plt.imshow(img)
plt.title("Montagna")
plt.show()
```

#Convert the image in gray_Scale and after apply Harris detection

```python
imgGray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(imgGray, cmap="gray")
plt.title("Gray Mountain")
plt.show()

Rfunction = cv.cornerHarris(imgGray,20,3,0.4)
print("Rmax=", Rfunction.max())

plt.imshow(Rfunction, cmap="jet")
plt.title("R matrix of image")
plt.show()
```

```python
img1 = img.copy()
img1[Rfunction>0.1*Rfunction.max()]=[255,0,0]
plt.imshow(img1)
plt.show()
```

#Apply the Shi-tomasi corner detection

```python
img = cv.imread('Montagna.jpg')
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
corners = cv.goodFeaturesToTrack(gray,15,0.01,10)
corners = np.int0(corners)
for i in corners:
    x,y = i.ravel()
    cv.circle(img,(x,y),3,255,-1)
plt.imshow(img),plt.show()
```

# Per quanto riguarda i risultati, credo che l'algoritmo Shi-Tomasi sia leggermente migliore dell'algoritmo Harris, in quanto in Shi-Tomasi, una volta inserita la soglia minima di accettazione, quest'ultimo scarta a priori tutti gli altri risultati che potrebbero non rappresentare degli angoli, mentre ciò avviene in maniera leggermente piu difficoltosa nell'algoritmo Harris. Ciò nonostante ho trovato molto efficaci entrambi.
